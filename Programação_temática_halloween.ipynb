```python
import random

def jogar_adivinhacao():
    """
    Fun√ß√£o principal do Jogo de Adivinha√ß√£o com tema Halloween.
    A bruxa guarda um n√∫mero secreto em seu caldeir√£o.
    """

    # 1. Configura√ß√£o do Jogo
    print("üéÉ Bem-vindo √† Cabana da Bruxa! üßô‚Äç‚ôÄÔ∏è")
    print("Para sair das garras da Bruxa, descubra o 'N√∫mero Secreto' em seu caldeir√£o.")
    print("Voc√™ tem poucas chances para descobrir, ou *UM FEITI√áO* cair√° sobre voc√™!\n")

    # Gera um n√∫mero secreto aleat√≥rio entre 1 e 50
    # (A Componente 'random' ser√° usada para gera√ß√£o de n√∫meros pseudo-aleat√≥rios.)
    numero_secreto = random.randint(1, 50)

    # Define o n√∫mero de tentativas
    tentativas_maximas = 5
    # Controle: Vari√°vel de controle que usar√° o 'break' para sair do loop.
    acertou = False
    tentativas_restantes = tentativas_maximas

    # 2. In√≠cio do Jogo
    # O conceito: 'while' loop para controlar o fluxo do jogo enquanto o jogador tiver chances e n√£o acertar.
    while tentativas_restantes > 0 and not acertou:
        print(f"\n‚ú® Tentativas restantes: {tentativas_restantes}")

        # Leitura da entrada do usu√°rio
        try:
            chute = int(input("üíÄ Qual o seu palpite entre 1 e 50? "))
        except ValueError:
            print("‚ö†Ô∏è Entrada inv√°lida! Digite apenas n√∫meros inteiros.\n")
            continue  # Volta para o in√≠cio do loop sem decrementar tentativas

        # 3. Valida√ß√£o da Entrada
        if chute < 1 or chute > 50:
            print("üö´ O palpite deve estar entre 1 e 50. Preste aten√ß√£o!")
            continue

        # 4. Verifica√ß√£o do Palpite (Estruturas Condicionais)
        tentativas_restantes -= 1

        if chute == numero_secreto:
            print(f"üéâ Parab√©ns! Voc√™ quebrou o feiti√ßo! O n√∫mero era {numero_secreto}!")
            acertou = True
        elif chute > numero_secreto:
            print("‚ò†Ô∏è Muito alto! O caldeir√£o borbulha dizendo que o n√∫mero √© *MENOR*...\n")
        else:
            print("üíÄ Muito baixo! A fuma√ßa do caldeir√£o indica que o n√∫mero √© *MAIOR*...\n")

    # 5. Fim do Jogo
    if not acertou:
        print("üëª Suas chances acabaram! O feiti√ßo est√° completo!")
        print(f"O N√∫mero Secreto da Bruxa era {numero_secreto}.")

if __name__ == "__main__":
    jogar_adivinhacao()
```

```python
import random
import os  # M√≥dulo para limpar a tela do console

def exibir_forca(erros):
    """Desenha a forca de acordo com o n√∫mero de erros."""
    # A forca √© um exemplo de como o algoritmo pode representar estados.
    stages = [
        """
        ------
        |    |
        |
        |
        |
        |
        =========
        """,
        # 1 erro
        """
        ------
        |    |
        |    O
        |
        |
        |
        =========
        """,
        # 2 erros
        """
        ------
        |    |
        |    O
        |    |
        |
        |
        =========
        """,
        # 3 erros
        """
        ------
        |    |
        |    O
        |   /|
        |
        |
        =========
        """,
        # 4 erros
        """
        ------
        |    |
        |    O
        |   /|\\
        |
        |
        =========
        """,
        # 5 erros (CORPO COMPLETO)
        """
        ------
        |    |
        |    O
        |   /|\\
        |   / \\
        |
        =========
        """
    ]
    print(stages[erros])

def jogar_forca():
    """Fun√ß√£o principal do Jogo da Forca com tema Halloween."""
    print("üï∏Ô∏èüéÉ Bem-vindo ao Jogo da Forca de Halloween! üßô‚Äç‚ôÄÔ∏è\n")
    print("Tente adivinhar a palavra MISTERIOSA (relacionada ao tema Halloween)")
    print("Mas cuidado... a cada erro a bruxa se aproxima at√© que o *FEITI√áO* se complete!\n")

    # 1. Configura√ß√£o (Lista de Palavras e Sorteio)
    palavras_tema = ["AB√ìBORA", "BRUXA", "MAGO", "VASSOURA", "FANTASMA", "ASSOMBRA√á√ÉO",
                     "CAVEIRA", "FEITI√áO", "ESQUELETO", "CALDEIR√ÉO", "MURCHO", "CONJURO"]
    palavra_secreta = random.choice(palavras_tema).upper()

    # 2. Vari√°veis de controle
    letras_descobertas = ["_" for letra in palavra_secreta]
    letras_erradas = []
    erros = 0
    max_erros = 5

    # 3. Loop Principal
    while erros < max_erros and "_" in letras_descobertas:
        os.system("cls" if os.name == "nt" else "clear")
        exibir_forca(erros)
        print("\nPalavra: ", " ".join(letras_descobertas))
        print(f"ü©∏ Letras erradas: {', '.join(letras_erradas)}")
        print(f"Tentativas restantes: {max_erros - erros}\n")

        # Entrada do jogador
        chute = input("üîÆ Escolha uma letra: ").upper()

        if len(chute) != 1 or not chute.isalpha():
            print("‚ö†Ô∏è Digite apenas uma letra v√°lida!\n")
            continue

        # 4. Verifica√ß√£o da letra
        if chute in letras_descobertas or chute in letras_erradas:
            print("‚ö†Ô∏è Voc√™ j√° tentou essa letra! Tente outra.\n")
            continue

        if chute in palavra_secreta:
            print("‚ú® Boa! A letra faz parte da palavra.\n")
            for i, letra in enumerate(palavra_secreta):
                if letra == chute:
                    letras_descobertas[i] = chute
        else:
            print("üíÄ A bruxa riu! Essa letra n√£o est√° na palavra...\n")
            letras_erradas.append(chute)
            erros += 1

    # 5. Encerramento do jogo
    os.system("cls" if os.name == "nt" else "clear")
    exibir_forca(erros)

    if "_" not in letras_descobertas:
        print(f"üéâ Parab√©ns! Voc√™ quebrou o feiti√ßo e descobriu a palavra: {palavra_secreta}!")
    else:
        print("üëª O feiti√ßo foi lan√ßado! A bruxa venceu desta vez...")
        print(f"A palavra secreta era: {palavra_secreta}")

if __name__ == "__main__":
    jogar_forca()
```
